### Train and test split

### Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
train_dataset = sc.fit_transform(train_dataset)
```

```python
train_dataset
```

```python
test_dataset = sc.transform(test_dataset)
```

```python
train_dataset
```

### Prepare the Data

```python
train_dataset.shape
```

```python
X_train = []
y_train = []
```

```python
for i in range(60,len(train_dataset)):
    X_train.append(train_dataset[i-60:i,:])
    y_train.append(train_dataset[i,:])
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

### Build the LSTM Architecture

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.layers import Input
```

```python
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional
```

```python
from tensorflow.keras import regularizers
```

### Init the model

```python
lstm_model = Sequential()
```

### Add input layer

```python
inp_layer = Input(shape = (60,1))
```

```python
lstm_model.add(inp_layer)
```

##### Add LSTM Layer -1

```python
# kernel_regularizer = regularizers.L2(0.001)
lstm_layer_1 = LSTM(units = 100, return_sequences = True, kernel_initializer = "random_uniform")
```

```python
bidirectional_lstm = Bidirectional(lstm_layer_1)
```

```python
lstm_model.add(bidirectional_lstm)
```

### Add a Dropout

```python
lstm_model.add(Dropout(0.1))
```

#### Add LSTM Layer -2

```python
lstm_layer_2 = LSTM(units = 100, return_sequences = False, kernel_initializer = "random_uniform", )
```

```python
bidirectional_lstm_2 = Bidirectional(lstm_layer_2)
```

```python
lstm_model.add(bidirectional_lstm_2)
```

#### Add Output layer

```python
output_layer = Dense(kernel_initializer = "random_uniform", activation = "linear", units = 1)
```

```python
lstm_model.add(output_layer)
```

```python
lstm_model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(lstm_model, show_shapes= True, )
```

```python
lstm_model.compile(loss = "mse", optimizer = "adam", metrics = "mse")
```

###### Train the model

```python
lstm_model.fit(X_train, y_train, batch_size = 32, epochs = 15)
```
