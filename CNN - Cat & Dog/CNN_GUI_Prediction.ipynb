```python

import numpy as np
from tkinter import *
from PIL import ImageTk, Image
from tkinter import filedialog
import os
from tensorflow.keras.models import load_model
import cv2

model = load_model('cnn_cd.h5')

#root -- Tkinter window
root = Tk()

#width*height
root.geometry("1000x500+100+100")
root.resizable(width=True, height=True)

def openfn():
    #open filedialog and return choosen filepath
    filepath = filedialog.askopenfilename(title='open')
    return filepath

def predict():
    
    filepath = openfn()
    
    
    image = cv2.imread(filepath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image,(128,128))
    image = image * (1.0/255.0)
    image = np.array([image])
    
    
    result = model.predict(image, verbose = 0)

    result = result > 0.5
    result = result.astype("int")[0][0]
    classes = ["Cat","Dog"]
    
    predicted_label = classes[result]
    
    #display the text on the UI
    label = Label(root, text = "Prediction : "+ predicted_label)
    #pack -- show the Label
    label.pack()
    
    #load the image in PIL
    img = Image.open(filepath)
    #resize the PIL image
    img = img.resize((250, 250), Image.ANTIALIAS)
    #Convert PIL into Tkinter format
    img = ImageTk.PhotoImage(img)
    
    #display the image on the UI
    panel = Label(root, image = img)
    panel.image = img
    #pack -- show the Label
    panel.pack()

#The window to show the button
#The text of th button
#command -- action to be performed after button click
btn = Button(root, text='Select Image', command = predict)
#pack -- show the button
btn.pack()

#create a loop
root.mainloop()
```
