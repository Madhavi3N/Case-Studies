# `Import required libraries`

```python
## Load the required Libraries

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.asst_node_Interactivity = 'all'
```

##  `Load the iris.csv dataset`

```python
df = pd.read_csv('data/Iris.csv')
```

```python
df
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
bar1=plt.bar(x='Iris-setosa', height=df['Species'].value_counts())
bar2=plt.bar(x='Iris-versicolor', height=df['Species'].value_counts())
bar3=plt.bar(x='Iris-virginica', height=df['Species'].value_counts())

plt.xlabel("Species")
plt.ylabel('Count')
plt.title("Frequency of the three Species")
  
plt.xticks(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.legend( (bar1, bar2, bar3), ('Iris-setosa', 'Iris-versicolor', 'Iris-virginica') )
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
plt.subplots(1,1,figsize=(8,4))
df['Species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Percentage of Iris Species")
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.pairplot(data=df,vars=['PetalLengthCm','PetalWidthCm','SepalLengthCm','SepalWidthCm'], hue='Species')
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(df['SepalLengthCm'],df['PetalLengthCm'],hue=df['Species'])
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x='Species',y='SepalLengthCm',data=df)
plt.subplot(2,2,2)
sns.boxplot(x='Species',y='SepalWidthCm',data=df)
plt.subplot(2,2,3)
sns.boxplot(x='Species',y='PetalLengthCm',data=df)
plt.subplot(2,2,4)
sns.boxplot(x='Species',y='PetalWidthCm',data=df)
plt.show()
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(data=df, y="Id", x="Species")
```

###### 7. Create a Violin Plot.

```python
### code here
sns.catplot(x='Species',y='SepalWidthCm',data=df,kind='violin')
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(data=df,vars=['PetalLengthCm','PetalWidthCm','SepalLengthCm'], hue='Species')
```

##### 9. Create a Heat Map

```python
### code here
sns.heatmap(df[['PetalLengthCm','SepalLengthCm']].corr(),annot=True)
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
sns.pairplot(data=df,vars=['PetalLengthCm','PetalWidthCm','SepalLengthCm','SepalWidthCm'], hue='Species',diag_kind="hist")
```
