```python
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=df);
plt.show()
```

```python
ordered_color = ['E','F','G','H','I','J']
sns.catplot(data = df, x = 'color', y = 'price', 
            order = ordered_color, color = 'purple', kind = 'box');
plt.show()
```

* Heatmap

```python
df_heatmap = df.pivot_table(values='carat',index='cut',columns='clarity',aggfunc=np.mean)
sns.heatmap(df_heatmap,annot=True)
plt.show()
```

#### Write a Python program to display all Multivariate plots from the dataframe.

```python
### code here
```

* Scatter plot or bubble chart with Regplot

```python
figure, ax = plt.subplots(figsize=(20,15))
sns.regplot(data = df,x = 'carat',y = "price",fit_reg = False, scatter_kws={'s':df['carat']*100}, logistic=True)
ax.set_xlabel("carat of daimonds")
ax.set_ylabel("price of diamonds")
ax.set_title("carat and price of Diamonds")
```

* Grouped Box Plot

```python
sns.boxplot(x='clarity',y='carat',hue='color',data=df)
```

* Heatmap

```python
sns.heatmap(data=df.corr(),annot=True)
```

* lmplot

```python
sns.lmplot(data=df, x="carat", y="price", col="cut", row="clarity", height=3, facet_kws=dict(sharex=False, sharey=False))
```

* Pair plot

```python
sns.pairplot(data=df,vars=['price','carat','depth','table'], hue='cut')
```

* Joint plot

```python
sns.jointplot(data = df, x = 'carat', y = 'price', kind = 'kde');
plt.show()
```
