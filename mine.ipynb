```python
import google.generativeai as genai
from PIL import Image
import matplotlib.pyplot as plt
```

```python
GOOGLE_API_KEY = "keys/gemini_api_key.txt"
f = open(GOOGLE_API_KEY)
key = f.read()
genai.configure(api_key=key)
```

```python
model = genai.GenerativeModel(model_name="gemini-1.5-flash")
```

```python
img = Image.open(r"C:\Users\91994\OneDrive\Pictures\Screenshots\Screenshot 2024-07-17 201759.png")
plt.imshow(img)
```

```python
prompt = "Describe the scene in the uploaded image in detail."
```

```python
res = model.generate_content([img,prompt])
print(res.text)
```

```python
prompt_1 = "Identify objects or obstacles within the image and highlight them, offering insights to enhance user thoughts and situational awareness."
```

```python
res1 = model.generate_content([img,prompt_1])
print(res1.text)
```
